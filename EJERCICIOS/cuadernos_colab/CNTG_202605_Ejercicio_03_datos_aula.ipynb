```python
import pandas as pd

URL = 'http://fpaniaguapython.github.io/datos/cntg-2026-05-26.csv'

dataframe = pd.read_csv(URL)

dataframe.info()
```

# Cambiar los nombres de las columnas

```python
dataframe.columns = ('timestamp','nombre','fecha_nacimiento','altura','peso','ingles','ciudad','ingresos', 'unnamed')
dataframe.info()
```

# Anonimizar los datos

```python
dataframe['nombre']='Anónimo'
dataframe['nombre']
```

# Reemplazar ' cm' por '' y ',' por '.'

```python
dataframe['altura'] = dataframe['altura'].str.replace(' cm','').str.replace(',','.')
#dataframe['altura'] = dataframe['altura'].str.replace(' cm','')
#dataframe['altura'] = dataframe['altura'].str.replace(',','.')
dataframe.head(10)
```

# Convertir a numérico la columna altura

```python
dataframe['altura'] = pd.to_numeric(dataframe['altura'], errors='coerce')
dataframe.info()
```

Filtrar las alturas menores de 3 (serían metros y habría que convertirlos a centímetros)

```python
dataframe.loc[dataframe['altura']<3, 'altura'] = dataframe.loc[dataframe['altura']<3]*100
```

# Reemplazar los NaN por la media de la altura

```python
media = dataframe['altura'].mean()
dataframe['altura'].fillna(media, inplace=True)
dataframe.head(10)
```

# Convertir tipo de la altura

```python
dataframe['altura'] = dataframe['altura'].astype('int64')
#dataframe.head(10)
dataframe.info()
```
