```python
import pandas as pd

URL='http://fpaniaguapython.github.io/datos/CCAA-ULTIMOS-DOCE-MESES-LAPTOP-H55DSDJ2.csv'

df = pd.read_csv(URL, sep=';', thousands=".") # OJO, indicamos que los . son separadores de miles
df.drop(0, inplace=True) # OJO, se pierde el índice 0
df.head()
```

```python
df.set_index('comunidad', inplace=True)
df.head()
```

```python
# Registros de comunidades que a lo largo del año han matriculado más de
# 250000 automóviles
#df['total']=df.iloc[:].sum(axis=1)
df['total']=df.sum(axis=1)
df.head()
mascara = df['total']>60000
df[mascara]
```

```python
# Seleccionar los registros(Comunidades) cuyo total esté entre 25000 y 50000
mascara = (df['total']>=25000) & (df['total']<=50000)
df[mascara]
#mascara = df['total'].between(25000,50000)
#df[mascara]
```
