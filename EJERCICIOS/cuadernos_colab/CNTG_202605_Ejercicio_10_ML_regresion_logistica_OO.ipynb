```python
from dataclasses import dataclass

import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

df = pd.read_csv('https://fpaniaguapython.github.io/datos/aves_extremadura.csv')

# longitud_cm	envergadura_cm	peso_g	longitud_pico_cm	migratoria
# habitat_bosque	habitat_campo_abierto	habitat_dehesa	habitat_humedal
# habitat_montana	habitat_ribera	habitat_urbano

@dataclass
class Ave:
  longitud_cm : int
  envergadura_cm: int
  peso_g: int
  longitud_pico_cm: float
  migratoria: int
  habitat_bosque: bool
  habitat_campo_abierto: bool
  habitat_dehesa: bool
  habitat_humedal: bool
  habitat_montana: bool
  habitat_ribera: bool
  habitat_urbano: bool
```

```python
print(df.info())
print(df.head())
```

```python
y = df['especie']
df.drop('especie', axis=1, inplace=True)
df.drop('nombre_cientifico', axis=1, inplace=True)
X = pd.get_dummies(df)
X.head()
```

```python
model = LogisticRegression()
model.fit(X, y)
'''
class Ave:
  longitud_cm : int
  envergadura_cm: int
  peso_g: int
  longitud_pico_cm: float
  migratoria: int
  habitat_bosque: bool
  habitat_campo_abierto: bool
  habitat_dehesa: bool
  habitat_humedal: bool
  habitat_montana: bool
  habitat_ribera: bool
  habitat_urbano: bool
'''

ave_desconocida_1 = Ave(longitud_cm=102, envergadura_cm=190, peso_g=3300,
                      longitud_pico_cm=18.5, migratoria=1,
                      habitat_bosque=False, habitat_campo_abierto=False,
                      habitat_dehesa=False, habitat_humedal=True,
                      habitat_montana=False, habitat_ribera=False,
                      habitat_urbano = False)

ave_desconocida_2 = Ave(longitud_cm=108, envergadura_cm=288,  peso_g=8900,
                      longitud_pico_cm=8.3, migratoria=0,
                      habitat_bosque=False, habitat_campo_abierto=False,
                      habitat_dehesa=False, habitat_humedal=True,
                      habitat_montana=False, habitat_ribera=False,
                      habitat_urbano = False)

ave_1 = list(ave_desconocida_1.__dict__.values())
ave_2 = list(ave_desconocida_2.__dict__.values())
aves = [ave_1, ave_2]
prob = model.predict_proba(aves)
clasificacion = model.predict(aves)

print("Probabilidad:", prob)
print("Clasificacion:", clasificacion)
print(model.classes_)
```
