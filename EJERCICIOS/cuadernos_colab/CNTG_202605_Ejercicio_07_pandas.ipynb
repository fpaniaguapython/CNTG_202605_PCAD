```python
import pandas as pd

df = pd.DataFrame({
    'id':(5,4,8,10),
    'lenguaje':('Python','Java','C','Cobol')
})
df.index = ['Cero','Uno','Dos','Tres']

df.head()
```

```python
# Acceso a la columna
df['lenguaje']
```

```python
# (LOC) Acceso a una fila por su índice
df.loc['Uno']
```

```python
# (ILOC) Acceso a una fila por su posición
df.iloc[0]
```

```python
# (ILOC) Acceso a un conjunto de filas por su posición con slicing
df.iloc[0:2]
```

```python
# (ILOC) Obtención de las dos últimas filas
df.iloc[-2:]
```

```python
# (ILOC) Obtención con slicing de combinaciones de filas y columnas
df.iloc[-2:,0:] # Dos últimas filas; Todas las columnas
```

```python
# Acceso a un conjunto de filas por su índice utilizando slicing
df.loc['Uno':'Tres']
```
