```python
from dataclasses import dataclass

import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

df = pd.read_csv('https://fpaniaguapython.github.io/datos/aves_extremadura.csv')
```

```python
print(df.info())
print(df.head())
```

```python
y = df['especie']
df.drop('especie', axis=1, inplace=True)
df.drop('nombre_cientifico', axis=1, inplace=True)
X = pd.get_dummies(df)
X.head()
```

```python
model = LogisticRegression()
model.fit(X, y)

# Opción 1 - Lista
# ave = [[102,	190,	3300,	18.5,	1,	False,	False,	False,	True,	False,	False,	False]]

# Opción 2 - Diccionario convertido a df
ave = pd.DataFrame([{
    'longitud_cm': 102,
    'envergadura_cm': 190,
    'peso_g': 3300,
    'longitud_pico_cm': 18.5,
    'migratoria': 1,
    'habitat_bosque': False,
    'habitat_campo_abierto': False,
    'habitat_dehesa': False,
    'habitat_humedal': True,
    'habitat_montana': False,
    'habitat_ribera': False,
    'habitat_urbano': False
}])

prob = model.predict_proba(ave)
clasificacion = model.predict(ave)

print("Probabilidad:", prob)
print("Clasificacion:", clasificacion)
print(model.classes_) # Muestra las opciones de la variable de salida
```
