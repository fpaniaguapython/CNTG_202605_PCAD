## Limpieza de un DF

```python
import pandas as pd
```

```python
clientes = pd.DataFrame({
  'nombre': ['Ana', 'Luis', 'Luis', None],
  'edad': [25, None, None, 40],
  'ciudad': ['Madrid ', 'Barcelona', 'Barcelona', 'Sevilla']
})
#clientes.drop_duplicates(inplace=True)
clientes = clientes.drop_duplicates()
clientes['edad'] = clientes['edad'].fillna(clientes['edad'].mean())
clientes['ciudad'] = clientes['ciudad'].str.strip()
clientes
```

## Función merge

```python
clientes = pd.DataFrame({
'id_cliente': [1, 2, 3, 4],
'nombre': ['Ana', 'Luis', 'Marta', 'Ramón']
})
ventas = pd.DataFrame({
'id_cliente': [1, 2, 3, 5],
'venta': [100, 200, 300, 800]
})
resultado = pd.merge(clientes, ventas, on='id_cliente') # Por defecto, inner
#resultado = pd.merge(clientes, ventas, on='id_cliente', how='inner') # Debe existir en los dos df
#resultado = pd.merge(clientes, ventas, on='id_cliente', how='left') # Toma todos los del primer df
#resultado = pd.merge(clientes, ventas, on='id_cliente', how='right') # Toma todos los del segundo df
#resultado = pd.merge(clientes, ventas, on='id_cliente', how='outer') # Toma todos los de los dos df

resultado
```

## Método join

```python
empleados = pd.DataFrame({
'nombre': ['Ana', 'Luis'],
'departamento': ['Ventas', 'Marketing']
}, index=[1, 2])

salarios = pd.DataFrame({
'salario': [3000, 4000]
}, index=[1, 2])

resultado = empleados.join(salarios)
resultado
```

## Pivot table (dataframe)

```python
ventas = pd.DataFrame(
    {
      'region': ['Norte', 'Norte', 'Sur', 'Sur','Norte', 'Norte'],
      'producto': ('A', 'B', 'A', 'B', 'A', 'B'),
      'ventas': [100, 150, 200, 250, 5, 7]
    }
)
ventas
```

## Método pivot_table (transformación)

```python
pivot = ventas.pivot_table(
  values='ventas',
  index='producto',
  columns='region',
  aggfunc='sum'
)
pivot
```

## Método pivot

```python
ventas = pd.DataFrame({
  'region': ['Norte', 'Norte', 'Sur', 'Sur','Este', 'Este'],
  'producto': ('A', 'B', 'A', 'B', 'A', 'B'),
  'ventas': [100, 150, 200, 250, 5, 7]
})
ventas
```

```python
pivot = ventas.pivot(
  index='region',
  columns='producto',
  values='ventas'
)
pivot
```
