```python
import pandas as pd

URL = 'http://fpaniaguapython.github.io/datos/cntg-2026-05-26.csv'

dataframe = pd.read_csv(URL)

dataframe.info()
```

```python
dataframe['Altura-2'] = dataframe['Altura'].astype(str)
dataframe['Altura-2'] = dataframe['Altura-2'].str.replace(',', '.', regex=False)
dataframe['Altura-2'] = dataframe['Altura-2'].str.extract(r'(\d+\.?\d*)').astype(float)

dataframe['Altura-2']
```
