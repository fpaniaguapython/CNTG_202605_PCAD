# Tratamiento y conversión de la columna Altura utilizando expresiones regulares

```python
import pandas as pd

URL = 'http://fpaniaguapython.github.io/datos/cntg-2026-05-26.csv'

dataframe = pd.read_csv(URL)

dataframe.info()
```

```python
dataframe['Altura'] = dataframe['Altura'].astype(str)
dataframe['Altura'] = dataframe['Altura'].str.replace(',', '.', regex=False)
dataframe['Altura'] = dataframe['Altura'].str.extract(r'(\d+\.?\d*)').astype(float)

dataframe['Altura']
```

# Multiplicar por 100 los valores menores de 3 (suponemos que dichos valores son alturas expresadas en metros y las queremos convertir a centímetros)

```python
dataframe.loc[dataframe['Altura']<3, 'Altura']=dataframe.loc[dataframe['Altura']<3, 'Altura']*100
dataframe['Altura'].fillna(dataframe['Altura'].mean(), inplace=True)
dataframe['Altura'] = dataframe['Altura'].astype(int)
dataframe
```
